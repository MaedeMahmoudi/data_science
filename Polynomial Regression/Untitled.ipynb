```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.pipeline import Pipeline
```

```python
#synthetic data
```

```python
np.random.seed(42)
```

```python
n_samples = 200
x = np.linspace(0,10,n_samples) #generate number with 10 distance between two samples
y = 0.5 * x**2 - 2*x + 3 + np.random.normal(0,2,n_samples)

x = x.reshape(-1,1)# convert to two dimention
```

```python
# data splitting
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2 , random_state = 42)
```

```python
#visualization
```

```python
plt.figure(figsize = (10,6))
plt.scatter(x_train , y_train , color = 'pink' , label = 'training data' , alpha = 0.7 )
plt.scatter(x_test , y_test , color = 'purple' , label = 'test data' , alpha = 0.7 )
plt.title('data for polynomial regression')
plt.xlabel('x attr')
plt.ylabel('goal y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#train
```

```python
degree = 2

model = Pipeline([
    ('poly' , PolynomialFeatures(degree = degree)),
    ('linear', LinearRegression())
])

model.fit(x_train , y_train)
```
