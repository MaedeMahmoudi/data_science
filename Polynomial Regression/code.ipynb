```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.pipeline import Pipeline
```

```python
#synthetic data
```

```python
np.random.seed(42)
```

```python
n_samples = 200
x = np.linspace(0,10,n_samples) #generate number with 10 distance between two samples
y = 0.5 * x**2 - 2*x + 3 + np.random.normal(0,2,n_samples)

x = x.reshape(-1,1)# convert to two dimention
```

```python
# data splitting
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2 , random_state = 42)
```

```python
#visualization
```

```python
plt.figure(figsize = (10,6))
plt.scatter(x_train , y_train , color = 'pink' , label = 'training data' , alpha = 0.7 )
plt.scatter(x_test , y_test , color = 'purple' , label = 'test data' , alpha = 0.7 )
plt.title('data for polynomial regression')
plt.xlabel('x attr')
plt.ylabel('goal y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#train
```

```python
degree = 2

model = Pipeline([
    ('poly' , PolynomialFeatures(degree = degree)),
    ('linear', LinearRegression())
])

model.fit(x_train , y_train)
```

```python
#Evaluation
```

```python
y_train_pred = model.predict(x_train)
y_test_pred = model.predict(x_test)

train_mse = mean_squared_error(y_train , y_train_pred)
test_mse = mean_squared_error(y_test , y_test_pred)

train_r2 = r2_score(y_train , y_train_pred)
test_r2 = r2_score(y_test , y_test_pred)

print(f'MSE for train data : {train_mse : .2f} , MSE for test data : {test_mse:.2f}')
print(f'R^2 for train data : {train_r2 : .2f} , R^2 for test data : {test_r2 : .2f}')
```

```python
#Visualization
```

```python
x_plot = np.linspace(0,10,100).reshape(-1 , 1)
y_plot = model.predict(x_plot)

plt.figure(figsize=(10,6))
plt.scatter(x_train , y_train , color = 'yellow' , label = 'train data' , alpha = 0.7)
plt.scatter(x_test , y_test , color = 'blue' , label = 'test data' , alpha = 0.7)
plt.plot(x_plot , y_plot , color = 'red' , linewidth = 2 ,label = 'predict model')
plt.title('second_degree Polynomial regression')
plt.xlabel('x attr')
plt.ylabel('goal y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Coef
```

```python
linear_reg = model.named_steps['linear']
poly_features = model.named_steps['poly']

print('model coef:')
for i, coef in enumerate(linear_reg.coef_):
    print(f'for x^{i}:{coef:.4f}')

print(f'intercept:{linear_reg.intercept_:.4f}')
```

```python
#Experiment with different degree
```

```python
degrees = [1, 2, 3, 5, 10]
plt.figure(figsize=(15, 10))

for i, degree in enumerate(degrees):  
    model = Pipeline([
        ('poly', PolynomialFeatures(degree=degree)),  
        ('linear', LinearRegression())
    ])
    model.fit(x_train, y_train)  

    X_plot = np.linspace(0, 10, 100).reshape(-1, 1)
    y_plot = model.predict(X_plot)

    y_pred = model.predict(x_test)  
    r2 = r2_score(y_test, y_pred)

    plt.subplot(2, 3, i+1)
    plt.scatter(x_train, y_train, color='pink', alpha=0.3, label='train data')  
    plt.scatter(x_test, y_test, color='purple', alpha=0.3, label='test data')  
    plt.plot(X_plot, y_plot, color='red', label=f'degree {degree}\nR²={r2:.2f}')
    plt.title(f'degree: {degree}')
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()
```
